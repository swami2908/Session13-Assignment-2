```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import datasets
```

##### Loading a dataset from pandas dataset and transforming into a pandas dataframe

```python
boston = datasets.load_boston()
features = pd.DataFrame(boston.data, columns=boston.feature_names)
targets = boston.target
features.head()
```

```python
# Checking Whether "targets" array have any NAN
np.isnan(targets).any()
```

```python
# DATA Preprocessing.
# Replacing NAN values with median
features.fillna(features.median,inplace=True)
```

```python
# Creating input in X and Y axis
X = features.iloc[:,:-1].values
Y = targets
print("X-DATASET:\n",X)
print("Y-DATASET:\n",Y)
```

```python
# Splitting the dataset into the Training set and Test set
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.10,random_state=42,shuffle=True)
print("X-TRAINING DATA:\n",X_train)
print("X-TEST DATA:\n",X_test)
print("Y-TRAINING DATA:\n",y_train)
print("Y-TEST DATA:\n",y_test)
```

```python
# Feature Scaling X data set
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
print("X-TRAINING DATA AFTER FEATURE SCALING:\n",X_train)
print("X-TEST DATA AFTER FEATURE SCALING:\n",X_test)
```

```python
# RANDOM FOREST Training / model fitting to the training data set
model = RandomForestRegressor(random_state=0, n_estimators=100)
model.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = model.predict(X_test)
y_pred
```

```python
model_score = model.score(X_train,y_train)
print("Co-Efficient of determination R^2 of the prediction.:",model_score)
print("Mean squared error: %.2f"% mean_squared_error(y_test, y_pred))
print('Test Variance score: %.2f' % r2_score(y_test, y_pred))
```

We can see that our R^2 score and MSE are both very good. This means that we have found a
best fitting model to predict the median price value of a house.
